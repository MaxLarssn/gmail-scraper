```python
# FTL Order Tracker

url = "https://accounts.google.com/signin/v2/identifier?service=mail&passive=true&rm=false&continue=https%3A%2F%2Fmail.google.com%2Fmail%2F&ss=1&scc=1&ltmpl=default&ltmplcache=2&emr=1&osid=1&flowName=GlifWebSignIn&flowEntry=ServiceLogin"
driver = uc.Chrome()
driver.get(url)

driver.find_element_by_xpath("//*[@id='identifierId']").send_keys(gmail)
driver.find_element_by_xpath("//*[@id='identifierNext']/div/button").click()
time.sleep(2)
driver.find_element_by_xpath("//*[@id='password']/div[1]/div/div[1]/input").send_keys(password)
driver.find_element_by_xpath("//*[@id='passwordNext']/div/button").click()

time.sleep(2)

url = f"https://mail.google.com/mail/u/0/?tab=rm&ogbl#search/{mailsearch}"
driver.get(url)

time.sleep(2)


driver.find_element_by_xpath('//*[@id=":1v"]').click()
time.sleep(1)

html = driver.page_source
soup = bs(html)

mails = [i.text for i in soup.select("span.g2")]
mails_lst = []
for i in mails:
    mails_lst.append(i + "@gmail ")
    
order = [i.text for i in soup.select("center > table > tbody > tr > td > table > tbody > tr:nth-child(5) > td > table > tbody > tr:nth-child(1) > td > table:nth-child(1) > tbody > tr > td > table > tbody > tr > td")]


order1 = []
for i in order:
    x = i.replace("[","").replace("'","").replace("]","").replace(",","").replace("   "," ").replace("   "," ").replace("\t","").replace("\n"," ").strip()
    order1.append(x)

order1 = str(order1)


order_number = []
for t in order1.split():
    if len(t) == 20:
        order_number.append(t)

        
time.sleep(5)

tracking_numbers = []
o = 0

while True:
    
    order_num = order_number[o]
    url = f"https://footlocker.narvar.com/footlocker/tracking/uk-mail?order_number={order_num}"
    driver.get(url)
    
    time.sleep(1)
    html = driver.page_source
    time.sleep(1)
    soup = bs(html)
    x = [i.text for i in soup.select("a.tracking-number")]
    x = str(x)
    x = x.replace("[", "").replace("]", "").replace("'", "")
    
    if x == "":
        x = "Not shipped"
        tracking_numbers.append(x)  
    else:
        tracking_numbers.append(x)
    o += 1
    
    if o == 3:
        driver.close()
        break       
```

```python
df = pd.DataFrame({"order number":order_number, "ups tracking number":tracking_numbers})
```
